```python
def GetMultpp900(multStart, multEnd):
    import ctypes
    fMultiCalib900 = ROOT.TFile("../../refs/MultCalibFT0_pp900Jun_LHC22cde_pass3.root", "READ")
    grCentVsMult = fMultiCalib900.Get("MultCalibFT0_pp900Jun_LHC22cde_pass3_pub_corr")
    intgr_err = ctypes.c_double(0)  # Create a ctypes double to store the error value
    intgr = grCentVsMult.IntegralAndError(grCentVsMult.GetXaxis().FindBin(multStart), grCentVsMult.GetXaxis().FindBin(multEnd), intgr_err, "")
    intgr /= np.double(multEnd - multStart)
    intgr_err /= np.double(multEnd - multStart)
    print(f"Integral from {multStart} to {multEnd} = {intgr} +- {intgr_err}")
    return [intgr, intgr_err]
```
